```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([1,2,3])
y=np.array([1,3,2])

plt.scatter(x,y,color="red")
```

```python
slope_current=0
intercept_current=0
learning_rate=0.01
iterations=1000

for i in range(iterations):
  y_predicted=slope_current*x+intercept_current

  slope_grad=-(2/3)*sum(x*(y-y_predicted))
  intercept_grad=-(2/3)*sum(y-y_predicted)
  slope_current=slope_current-learning_rate*slope_grad
  intercept_current=intercept_current-learning_rate*intercept_grad
  plt.plot(x,y_predicted)
  
plt.scatter(x,y,color="red")
plt.plot(x,y_predicted,color="blue")
plt.show()
```
